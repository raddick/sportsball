```python
with open('events1927and1938.json','r') as f:
    allgames = json.load(f)
print('ok')
```

```python
allgames_df = pandas.DataFrame.from_dict(allgames)
column_list = ['event_in_game', 'event_type', 'inning', 'home_or_visitor', 'batterID']
column_list += ['pitch_count', 'pitch_sequence', 'event', 'comment', 'sub_batting_order', 'sub_position', 'gameID']

events_df = pandas.DataFrame(columns=column_list)

for thisgameindex, row in allgames_df.iterrows():
    this_game_events_df = pandas.DataFrame(row['events'])
    this_game_events_df = this_game_events_df.append(row['id'], axis=1)
    this_game_events_df.columns = column_list
    for i, thisevent in this_game_events_df.iterrows():
        if (thisevent['comment'] != ''):
            this_game_events_df['comment'].iloc[i-1] += thisevent['comment']
    this_game_events_df = this_game_events_df[this_game_events_df['event_type'] != 'com']
    this_game_events_df = this_game_events_df.drop('event_in_game', axis=1)
    this_game_events_df = this_game_events_df.reset_index(drop=True)
    this_game_events_df.index.name = 'event_in_game'
    this_game_events_df['gameID'] = thisgameindex
    this_game_events_df = this_game_events_df.reset_index()
    events_df += this_game_events_df
events_df.index.name = 'eventID'
print('done')
```
